```python
class Peghole:
    def __init__(self,idex,jdex, colour, window):
        self.i = idex
        self.j = jdex
        self.color = colour
        self.canvas = window
        self.scanned = False
    def show(self, xpos, ypos, size):
        if self.scanned == False : 
            pygame.draw.circle(self.canvas, self.color , ( xpos,ypos ) , size)
```

```python
class Board:
    def __init__(self,dimension, width, height , window):
        self.dim = dimension
        self.screenwidth = width
        self.screenheight = height
        self.canvas = window
        self.Pegstore = np.empty(self.dim*self.dim, dtype=object)
        self.PegInit()
    def PegInit(self):
        for i in range(self.dim):
            for j in range(self.dim):
 #               x = Rx*(1+i+j)
 #               y = self.screenheight*0.5+Ry*(i-j)
                self.Pegstore[i+j*Dim] = Peghole(i,j,empty,self.canvas)
    def show(self):
        Rx = self.screenwidth/(2*(self.dim))
        Ry = Rx*0.5
        for i in range(self.dim):
            for j in range(self.dim):
                x = Rx*(1+i+j)
                y = self.screenheight*0.5+Ry*(i-j)
                self.Pegstore[i+j*Dim].show(x,y,Ry*0.5)
#    def prompt(self):
#        BoardToAI = []
                                    
```

```python
class agent:
    def __init__(self, dimension, name1, name2, Pegboard, cyborg1 , cyborg2):
        self.Dim = dimension
        self.player = [name1,name2]
        self.turn = False
        self.playerColors = [red, blue]
        self.Pegfield = Pegboard
        self.KeyInput = ""
        self.Pegroad = []
        self.win = False
        self.isHuman = [cyborg1 , cyborg2] # if cyborg - player is computer! - if False Human
    def switchPlayerColors(self):
        self.player = self.player[::-1]
    def nextTurn(self):
        self.turn = not self.turn
    def turnOf(self):
        return self.player[self.turn]
#    def takesTurn(name, i, j):
#        if name == self.turnOf() and i <= self.Dim and j <= self.Dim and self.Pegfield[i,j].color == empty:
#            self.Pegfield[i,j].color = self.playerColors[self.turn]
#            self.nextTurn()
#            return True
#        else:
#            return False
    def validMove(self, spot) :
        if self.Pegfield[spot].color == empty:
            return true
    def handleKeyInput(self, string, ENTER):
        if ENTER:
            #TEST
            self.Pegfield[int(self.KeyInput)].color = red
            print(self.KeyInput)
            print("p")
            self.KeyInput = ""
        else:
            self.KeyInput += string
    def checkForWin(self) : 
        self.Pegroad = []
        if self.turn:
            for P in range(Dim) : 
                self.Pegroad.append(P)
        else : 
            for P in range(Dim) :
                self.Pegroad.append(P*Dim)
        while self.Pegroad :
 #  #         print(self.Pegroad)
            if self.Pegfield[self.Pegroad[0]].color == self.playerColors[self.turn] and self.Pegfield[P].scanned == False :
                self.Perculation(self.Pegroad[0]) #THIS WONT HIT (Only in a 1x1 Board) BUT THE LATTER ONE WILL...
                self.Pegfield[self.Pegroad[0]].scanned = True   # color error
                self.scanNeighbours(self.Pegroad[0])
#                print(self.Pegroad)
#                time.sleep(1)
#                self.Pegfield[self.Pegroad[0]].scanned = False
            del self.Pegroad[0]
        for P in range(Dim*Dim) :
            self.Pegfield[P].scanned = False
    def scanNeighbours(self, P) :
        Neighbours = []
        if P%Dim :
            Neighbours = [P-1,P+1,P+Dim,P-Dim,P+Dim-1,P-Dim+1]
        else : 
            Neighbours = [P+1,P+Dim,P-Dim,P-Dim+1]
        for I in Neighbours :
            if I >= 0 and I < Dim*Dim :
    #            print(self.Pegroad)
                if self.Pegfield[I].color == self.playerColors[self.turn] and  self.Pegfield[I].scanned == False :
                    self.Perculation(P) #THIS WILL HIT
                    self.Pegroad.append(I)
    #                print(self.Pegroad)
    def Perculation(self, P) :
        if self.turn:
            if P >= Dim*Dim-Dim :
                self.win = True
        else :
            for J in range(Dim):
                if P == J*Dim+Dim-1 :
                    self.win = True
```

```python
tf.random.set_seed(1234) # for consistent results
model = Sequential(
    [              
        ### START CODE HERE ###
        tf.keras.Input(shape=(Dim*Dim,)),
        Dense(25, activation = 'relu', name = "L1"),
        Dense(15, activation = 'relu', name = "L2"),
        Dense(Dim*Dim, activation = 'linear', name = "L3"),
    
    ], name = "my_model"
)


model.compile(
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
)


history = model.fit(
    X,y,
    epochs=40
)
```
